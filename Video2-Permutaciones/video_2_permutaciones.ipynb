```python
%%manim -qh -v WARNING Video2Escena2Problema

class Video2Escena2Problema(Scene):
    def construct(self):
        self.mostrar_escena_inicial()
        self.mostrar_pregunta()
        
        self.animar_niños_a_vagon()
        self.mostrar_asiento_1()
        self.iluminar_asientos_2_3()
        self.mostrar_asiento_2()
        self.mostrar_asiento_3()
        
        self.crear_arbol_susana()
        
        self.remove(self.vagon, self.susana, self.diana, self.alex)
        self.mostrar_titulos_asientos()
        self.clear()

    def mostrar_escena_inicial(self):
        self.montaña_rusa = SVGMobject("imagenes/atracciones/montaña_rusa.svg" \
                                      ).scale(2.9).move_to([0, .5, 0])
        self.vagon = SVGMobject("imagenes/vagon_permutaciones/vagon.svg" \
                               ).scale(1).move_to([3, -2, 1])
        
        self.susana, self.susana_texto = self.crear_niñx_nombre(
            ruta_imagen="imagenes/niñxs/niña_verde.svg", pos_imagen=[-5, -2, 0], 
            nombre="Susana", pos_texto=[-5, -0.5, 0]
        )
        self.diana, self.diana_texto = self.crear_niñx_nombre(
            ruta_imagen="imagenes/niñxs/niña_naranja.svg", pos_imagen=[-3.5, -2, 0], 
            nombre="Diana", pos_texto=[-3.5, -0.5, 0]
        )
        self.alex, self.alex_texto = self.crear_niñx_nombre(
            ruta_imagen="imagenes/niñxs/niño_amarillo.svg", pos_imagen=[-2, -2, 0], 
            nombre="Alex", pos_texto=[-2.1, -0.5, 0]
        )
        
        self.montaña_rusa.set_fill(opacity=0.4).set_stroke(opacity=0.2)

        self.play(FadeIn(self.montaña_rusa))
        self.play(
            FadeIn(self.susana, self.diana, self.alex, self.vagon),
            run_time=1
        )
        self.wait(1)
    
    def crear_niñx_nombre(self, ruta_imagen, pos_imagen, nombre, pos_texto):
        niñx = SVGMobject(ruta_imagen).scale(1.2).move_to(pos_imagen)
        texto = Tex(nombre, font_size=20).move_to(pos_texto).scale(1.2)
        
        return niñx, texto

    def mostrar_pregunta(self):
        pregunta = Tex(
            r"\text{¿De cuántas maneras distintas pueden sentarse} \\ "
            r"\text{en el vagón para probar todos los} \\ "
            r"\text{distintos puntos de vista?}",
            font_size=50
        )
        pregunta.move_to([0, 2.6, 0])
        
        self.play(Write(pregunta), FadeOut(self.montaña_rusa))
        self.wait(2)
        self.play(FadeOut(pregunta))

    def animar_niños_a_vagon(self):
        niñxs_texto = Tex(r"\text{Son 3 amigos y} \\ \text{hay 3 asientos}"
                          r" \\ \text{disponibles.}", font_size=50)
        niñxs_texto.move_to([-4.5, 2.6, 0])
        
        self.play(
            self.susana.animate.move_to([-1.6, 3, 0]).scale(0.65),
            self.susana_texto.animate.move_to([-0.6, 3, 0]).scale(1.2),
            
            self.diana.animate.move_to([-1.6, 1.3, 0]).scale(0.65),
            self.diana_texto.animate.move_to([-0.6, 1.3, 0]).scale(1.2),
            
            self.alex.animate.move_to([-1.6, -0.4, 0]).scale(0.65),
            self.alex_texto.animate.move_to([-0.6, -0.4, 0]).scale(1.2),
            
            self.vagon.animate.move_to([0, -2.5, 0]).scale(1.3)
        )
        
        self.play(Write(niñxs_texto))
        self.wait(0.5)
        self.play(FadeOut(niñxs_texto))
        
    def mostrar_asiento_1(self):
        asiento_1_texto = Tex(r"\text{3 personas} \\ "
                              r"\text{elegibles para} \\ "
                              r"\text{el asiento 1.}", font_size=50)
        vagon_1 = SVGMobject("imagenes/vagon_permutaciones/vagonL1.svg" \
                            ).move_to([0,-2.5,0]).scale(1.358).set_fill(opacity=1)
        numero_1 = Integer(number=1).scale(2).move_to([-1.5,-1.8,0]
                                                     ).set_color(BLACK) 
        
        asiento_1_texto.move_to([-4.5, 2.6, 0])
        
        self.add(vagon_1,numero_1)
        self.play(Write(asiento_1_texto))
        self.wait(.5)
        
        self.remove(vagon_1, numero_1)
        self.add(self.vagon)
        self.play(FadeOut(self.diana_texto, self.alex_texto))
        self.wait(.5)
        
        self.play(
            FadeOut(asiento_1_texto),
            FadeOut(self.susana_texto),
            self.diana.animate.move_to([.3,2.6,0]).scale(1.4),
            self.alex.animate.move_to([.3,0,0]).scale(1.3),
            self.susana.animate.move_to([-1.6,-1.5,-1]).scale(1.4)
        )
        self.wait(.5)
        
    def iluminar_asientos_2_3(self):
        vagon_2_3 = SVGMobject("imagenes/vagon_permutaciones/vagonL2L3.svg" \
                              ).move_to([0,-2.5,0]).scale(1.363).set_fill(opacity=1)
        asiento_2_3_texto = Tex(r"\text{Quedan 2 personas y } \\ "
                                r"\text{2 asientos disponibles.}", font_size=50)
        
        asiento_2_3_texto.move_to([-4, 2.6, 0])
        self.add(vagon_2_3)
        self.play(Write(asiento_2_3_texto))
        self.wait(.5)
        
        self.remove(vagon_2_3)
        self.play(FadeOut(asiento_2_3_texto))
        

    def mostrar_asiento_2(self):
        vagon_2 = SVGMobject("imagenes/vagon_permutaciones/vagonL2.svg" \
                            ).move_to([0,-2.5,0]).scale(1.363).set_fill(opacity=1)
        asiento_2_texto = Tex(r"\text{2 personas} \\ \text{elegibles para}"
                              r" \\ \text{el asiento 2.}", font_size=50)
        numero_2 = Integer(number=2).scale(2).move_to([.4,-1.8,0]).set_color(BLACK)
        
        asiento_2_texto.move_to([-4.5, 2.6, 0])
        self.add(vagon_2, numero_2)
        self.play(Write(asiento_2_texto))
        self.wait(1)
        
        self.remove(vagon_2, numero_2)
        self.play(
            FadeOut(asiento_2_texto), 
            self.diana.animate.move_to([2.1,.5,0]).scale(1),
            self.alex.animate.move_to([.3,-1.7,0]).scale(1.1)
        )
        self.wait(.5)
    
    def mostrar_asiento_3(self):
        vagon_3 = SVGMobject("imagenes/vagon_permutaciones/vagonL3.svg" \
                            ).move_to([0,-2.5,0]).scale(1.347).set_fill(opacity=1)
        asiento_3_texto = Tex(r"\text{1 personas} \\ \text{elegibles para}"
                              r" \\ \text{el asiento 3.}", font_size=50)
        numero_3 = Integer(number=3).scale(2).move_to([2.2,-1.8,0]).set_color(BLACK)
        
        asiento_3_texto.move_to([-4.5, 2.6, 0])
        self.add(vagon_3, numero_3)
        self.play(Write(asiento_3_texto))
        self.wait(1)
        
        self.remove(vagon_3, numero_3)
        self.play(
            FadeOut(asiento_3_texto), 
            self.diana.animate.move_to([2.1,-1.7,0]).scale(1.1)
        )
        self.wait(1.2)
        
    def crear_arbol_susana(self):
        susana_1 = SVGMobject("imagenes/niñxs/niña_verde.svg" \
                             ).scale(0.8).move_to([-4, 2, 0])
        self.arbol_susana = VGroup()
        
        self.arbol_parte_1(susana_1)
        self.cambiar_diana_alex()
        self.arbol_parte_2(susana_1)

    def arbol_parte_1(self, susana_1):
        alex_1 = SVGMobject("imagenes/niñxs/niño_amarillo.svg" \
                           ).scale(0.8).move_to([-1, 3, 0])
        diana_1 = SVGMobject("imagenes/niñxs/niña_naranja.svg" \
                            ).scale(0.85).move_to([2, 3, 0])
        vagon_1_3_2 = SVGMobject("imagenes/vagon_permutaciones/vagon132.svg" \
                                ).scale(0.7).move_to([5, 3, 0])

        arista_1 = self.crear_arista(susana_1, alex_1)
        arista_2 = self.crear_arista(alex_1, diana_1)
        arista_3 = self.crear_arista(diana_1, vagon_1_3_2)
        
        self.arbol_susana.add(
            susana_1, diana_1, alex_1,
            arista_1, arista_2, arista_3,
            vagon_1_3_2
        )
        
        self.play(FadeIn(susana_1))
        self.play(Create(arista_1), FadeIn(alex_1))
        self.play(Create(arista_2), FadeIn(diana_1))
        self.play(Create(arista_3), FadeIn(vagon_1_3_2))
        
    def cambiar_diana_alex(self):
        self.play(
            self.diana.animate.move_to([2.1, 0.5, 0]),
            self.alex.animate.move_to([0.3, 0, 0])
        )
        self.play(
            self.diana.animate.move_to([0.34, -1.5, 0]),
            self.alex.animate.move_to([2.2, -1.79, 0])
        )
    
    def arbol_parte_2(self, susana_1):
        alex_2 = SVGMobject("imagenes/niñxs/niño_amarillo.svg").scale(0.8).move_to([2, 1, 0])
        diana_2 = SVGMobject("imagenes/niñxs/niña_naranja.svg").scale(0.85).move_to([-1, 1, 0])
        vagon_1_2_3 = SVGMobject("imagenes/vagon_permutaciones/vagon123.svg").scale(0.7).move_to([5, 1, 0])

        arista_4 = self.crear_arista(susana_1, diana_2)
        arista_5 = self.crear_arista(diana_2, alex_2)
        arista_6 = self.crear_arista(alex_2, vagon_1_2_3)
        
        self.arbol_susana.add(alex_2, diana_2, arista_4, arista_5, arista_6, vagon_1_2_3)
        
        self.play(Create(arista_4), FadeIn(diana_2))
        self.play(Create(arista_5), FadeIn(alex_2))
        self.play(Create(arista_6), FadeIn(vagon_1_2_3))

    def mostrar_titulos_asientos(self):
        asiento_1_texto = Tex(r"\text{Asiento} \\ \text{1}", font_size=50).move_to([-5.35, 2.6, 0])
        asiento_2_texto = Tex(r"\text{Asiento} \\ \text{2}", font_size=50).move_to([-2.25, 2.6, 0])
        asiento_3_texto = Tex(r"\text{Asiento} \\ \text{3}", font_size=50).move_to([1.05, 2.6, 0])
        self.play(
            self.arbol_susana.animate.move_to([0, -1, 0]).scale(1.07),
            FadeIn(
                asiento_1_texto, 
                asiento_2_texto, 
                asiento_3_texto
            )
        )
        self.wait(2)

    def crear_arista(self, start_obj, end_obj):
        linea = Line(start=start_obj.get_center(), end=end_obj.get_center()).set_z_index(-1)
        #edge.z_index = -1
        return linea
```

```python
%%manim -qh -v WARNING Video2Escena3Arbol

class Video2Escena3Arbol(Scene):
    def construct(self):
        self.mostrar_asientos()
        self.crear_arbol_con_asientos()
        
        self.mostrar_principio_producto()
        self.mostrar_operaciones()
        
        self.animar_movimientos_asientos()
        self.esperar_y_despedir()

    def mostrar_asientos(self):
        self.asiento_1_texto = self.crear_asiento_texto(1, [-2.1, 3, 0])
        self.asiento_2_texto = self.crear_asiento_texto(2, [0.1, 3, 0])
        self.asiento_3_texto = self.crear_asiento_texto(3, [2.3, 3, 0])
        
        self.play(FadeIn(
            self.asiento_1_texto, 
            self.asiento_2_texto, 
            self.asiento_3_texto
        ))

    def crear_asiento_texto(self, numero, posicion):
        asiento_texto = Tex(rf"\text{{Asiento}} \\ \text{{{numero}}}" \
                   , font_size=50).move_to(posicion)
        
        return asiento_texto

    def crear_arbol_con_asientos(self):
        arbol_niveles, self.arbol, _, _ = construir_arbol_vagon(1)
        self.arbol.move_to([0, -0.8, 0])
        
        crear_arbol_niveles(self, arbol_niveles)
        self.wait(2)
        
        self.arbol.add(self.asiento_1_texto, self.asiento_2_texto, self.asiento_3_texto)
        self.play(self.arbol.animate.scale(.95).move_to([1.2, -.3, 0]))

    def mostrar_principio_producto(self):
        regla_producto_texto = Tex(r"\text{Principio del} \\ "
                                   r" \text{Producto}", font_size=60)
        regla_producto_texto.move_to([-4.5, 3, 0])
        
        self.play(Write(regla_producto_texto))

    def mostrar_operaciones(self):
        self.mostrar_ovalos_y_numeros()

    def mostrar_ovalos_y_numeros(self):
        ovalo1 = self.crear_ovalo(posicion=[-.78, -0.9, 0])
        ovalo2 = self.crear_ovalo(posicion=[1.3, -0.95, 0], altura=1.9)
        ovalo3 = self.crear_ovalo(posicion=[3.4, -0.46, 0], altura=1)
        
        numero1 = Integer(number=3, color=WHITE).scale(1.5).move_to([-6, 1, 0])
        numero2 = Integer(number=2, color=WHITE).scale(1.5).move_to([-5, 1, 0])
        numero3 = Integer(number=1, color=WHITE).scale(1.5).move_to([-4, 1, 0])
        resultado = Tex("= 6", font_size=80).next_to([-4, 1, 0], RIGHT)
        
        self.play(Write(ovalo1))
        self.play(Transform(ovalo1, numero1))
        self.mostrar_simbolo_multiplicacion(posicion=[-5.5, 1, 0])

        self.play(Write(ovalo2))
        self.play(Transform(ovalo2, numero2))
        self.mostrar_simbolo_multiplicacion(posicion=[-4.5, 1, 0])

        self.play(Write(ovalo3))
        self.play(Transform(ovalo3, numero3))
        self.play(FadeIn(resultado))

    def crear_ovalo(self, posicion, altura=5.2):
        ovalo =  Ellipse(width=1.5, height=altura, color=BLUE, 
                          fill_opacity=0, stroke_width=5).move_to(posicion)
        
        return ovalo

    def mostrar_simbolo_multiplicacion(self, posicion):
        multiplicador = Tex("x", font_size=80).move_to(posicion)
        
        self.play(FadeIn(multiplicador))

    def animar_movimientos_asientos(self):
        posiciones = [
            [1.295, 2.662, 0],
            [-0.795, 2.662, 0],
            [3.385, 2.662, 0]
        ]

        self.play(
            self.asiento_1_texto.animate.move_to(posiciones[0]),
            self.asiento_2_texto.animate.move_to(posiciones[1])
        )
        self.play(
            self.asiento_1_texto.animate.move_to(posiciones[2]),
            self.asiento_3_texto.animate.move_to(posiciones[0])
        )
        self.play(
            self.asiento_3_texto.animate.move_to(posiciones[1]),
            self.asiento_1_texto.animate.move_to(posiciones[0]),
            self.asiento_2_texto.animate.move_to(posiciones[2])
        )

    def obtener_asientos_texto(self):
        asiento_1_texto = self.crear_asiento_texto("Asiento 1", [-2.1, 3, 0])
        asiento_2_texto = self.crear_asiento_texto("Asiento 2", [0.1, 3, 0])
        asiento_3_texto = self.crear_asiento_texto("Asiento 3", [2.3, 3, 0])
        
        return asiento_1_texto, asiento_2_texto, asiento_3_texto

    def esperar_y_despedir(self):
        self.wait(2)
        self.play(FadeOut(self.arbol))
        self.wait(2)
```

```python
%%manim -qh -v WARNING Video2Escena4Factorial

class Video2Escena4Factorial(Scene):
    def construct(self):
        self.mostrar_escena_pasada()
        self.factorial_3()
        
        self.factorial()
        self.cambiar_opacidad_2()
        self.cambiar_opacidad_3()
        self.cambiar_opacidad_4()
        self.cambiar_opacidad_5()
        self.opacidad_1()
        self.numeros.add(self.igual_5, self.n_factorial)
        for factorial in self.factoriales:
            self.numeros.add(factorial)
        
        self.igualdad()
        self.clear()
        
    def mostrar_escena_pasada(self):
        numero_3 = Integer(number=3, color = WHITE).scale(1.5).move_to([-6,1,0])
        numero_2 = Integer(number=2, color = WHITE).scale(1.5).move_to([-5,1,0])
        numero_1 = Integer(number=1, color = WHITE).scale(1.5).move_to([-4,1,0])
        self.regla_prod_texto = Tex(r"\text{Principio del} \\ "
                               r"\text{Producto}",font_size = 60).move_to([-4.5, 3, 0])
        multiplicador_1 = Tex("$\\times$",font_size = 80).move_to([-5.5,1,0])
        multiplicador_2 = Tex("$\\times$",font_size = 80).move_to([-4.5,1,0])
        self.igual_6 = Tex("$= 6$",font_size = 80).next_to(numero_1, RIGHT)
        
        self.factorial_3_grupo = VGroup()
        self.factorial_3_grupo.add(numero_3, numero_2, numero_1, 
                                   multiplicador_1, multiplicador_2)
        
        self.add(self.regla_prod_texto, numero_3, numero_2, numero_1, 
                 multiplicador_1, multiplicador_2, self.igual_6)
        self.wait(2)
    
    def factorial_3(self):
        factorial_texto = Tex(r"\text{Factorial}",
                              font_size = 60).move_to([0,3,0])
        igual = Tex("$=$",font_size = 80).move_to([-1.8,.94,0])
        factorial_3 = Tex("$3!$",font_size = 80).move_to([-1.0,1,0])
        flecha = Arrow([-.77, 2.95, 0], [-.77, 1.1, 0])
        
        self.play(FadeIn(igual, factorial_3))
        self.wait(2)
        
        self.play(Write(factorial_texto), Write(flecha))
        self.wait(2)
        self.play(FadeOut(self.regla_prod_texto, self.igual_6))
        self.play(
            igual.animate.move_to([-.16,1,0]),
            self.factorial_3_grupo.animate.move_to([1.45,1,0])
        )
        
        self.factorial_3_grupo.add(factorial_3, igual)
        self.wait(2)
        
        self.play(
            factorial_texto.animate.scale(1.3).move_to([-4.5,3,0]), 
            FadeOut(flecha, self.factorial_3_grupo)
        )
        self.wait(2)
        
    def factorial(self):
        self.n_factorial = Tex("$n! = n \\times (n-1) \\times \\cdot "
                          "\\cdot \\cdot \\times 2 \\times 1$",font_size = 80 \
                         ).move_to([-0.8,2,0])
        
        self.crear_factoriales()
        self.play(Write(self.n_factorial))
        
        for factorial in self.factoriales:
            self.play(Write(factorial))
            self.wait(.5)
        self.wait(1)
        
    def crear_factoriales(self):
        self.factoriales = [
            Tex("$1! = 1$", 
                font_size=80).move_to([-4.325, 1, 0]),
            
            Tex("$2! = 2 \\times 1$", 
                font_size=80).move_to([-3.625, 0, 0]),
            
            Tex("$3! = 3 \\times 2 \\times 1$", 
                font_size=80).move_to([-2.911, -1, 0]),
            
            Tex("$4! = 4 \\times 3 \\times 2 \\times 1$", 
                font_size=80).move_to([-2.205, -2, 0]),
            
            Tex("$5! = 5 \\times 4 \\times 3 \\times 2 \\times 1$", 
                font_size=80).move_to([-1.48, -3, 0])
        ]
    
    def cambiar_opacidad_2(self):
        self.igual_2 = Tex("$= 2 \\times$",font_size = 80).move_to([-.8,0,0])
        factorial_1 = Tex("$1!$",font_size = 80).move_to([.5,0,0])
        
        self.cambiar_opacidad(indice_transparencia=1)
        self.factorial_1_igual = copy.deepcopy(self.factoriales[0])
        
        self.add(self.factorial_1_igual)
        self.play(FadeIn(self.igual_2))
        self.play(Transform(self.factorial_1_igual, factorial_1), run_time = 1)
        self.wait(1)
        
        self.play(self.factoriales[1].animate.set_opacity(.2),
                  self.factorial_1_igual.animate.set_opacity(.1),
                  self.igual_2.animate.set_opacity(.2),
                  self.factoriales[2].animate.set_opacity(1))
        
    def cambiar_opacidad_3(self):
        self.factorial_2_igual = copy.deepcopy(self.factoriales[1])
        self.igual_3 = Tex("$= 3 \\times$",font_size = 80).move_to([.6,-1,0])
        factorial_2 = Tex("$2!$",font_size = 80).move_to([1.9,-1,0])
        
        self.play(FadeIn(self.igual_3))
        self.play(Transform(self.factorial_2_igual, factorial_2), run_time = 1)
        self.wait(1)
        
        self.play(self.factoriales[2].animate.set_opacity(.2),
                  self.factorial_2_igual.animate.set_opacity(.075),
                  self.igual_3.animate.set_opacity(.2),
                  self.factoriales[3].animate.set_opacity(1))
        
    def cambiar_opacidad_4(self):
        self.factorial_3_igual = copy.deepcopy(self.factoriales[2])
        self.igual_4 = Tex("$= 4 \\times$",font_size = 80).move_to([2,-2,0])
        factorial_3 = Tex("$3!$",font_size = 80).move_to([3.3,-2,0])
        
        self.play(FadeIn(self.igual_4))
        self.play(Transform(self.factorial_3_igual, factorial_3), run_time = 1)
        self.wait(1)
        
        self.play(self.factoriales[3].animate.set_opacity(.2),
                  self.factorial_3_igual.animate.set_opacity(.05),
                  self.igual_4.animate.set_opacity(.2),
                  self.factoriales[4].animate.set_opacity(1))
        
    def cambiar_opacidad_5(self):
        self.factorial_4_igual = copy.deepcopy(self.factoriales[3])
        self.igual_5 = Tex("$= 5 \\times$",font_size = 80).move_to([3.5,-3,0])
        factorial_4 = Tex("$4!$",font_size = 80).move_to([4.8,-3,0])
        
        self.play(FadeIn(self.igual_5))
        self.play(Transform(self.factorial_4_igual, factorial_4), run_time = 1)
        self.wait(2)
        
    def cambiar_opacidad(self, indice_transparencia):
        self.play(*[factorial.animate.set_opacity(.2) 
                    for i, factorial in enumerate(self.factoriales) if i != indice_transparencia])
        
    def opacidad_1(self):
        
        self.play(*[self.factorial.animate.set_opacity(1) 
                    for i, self.factorial in enumerate(self.factoriales) if i != 5],

                  self.factorial_1_igual.animate.set_opacity(1),
                  self.igual_2.animate.set_opacity(1),
                  self.factorial_2_igual.animate.set_opacity(1),
                  self.igual_3.animate.set_opacity(1),
                  self.factorial_3_igual.animate.set_opacity(1),
                  self.igual_4.animate.set_opacity(1),
                  self.factorial_4_igual.animate.set_opacity(1),
            )
        self.numeros = VGroup(
            self.igual_2, self.factorial_1_igual, 
            self.igual_3, self.factorial_2_igual, 
            self.igual_4, self.factorial_3_igual,
            self.factorial_4_igual
        )

    def igualdad(self):
        valor_factorial_2 = Tex("$= 2$",font_size = 80).move_to([1.65,0,0]).set_opacity(0)
        valor_factorial_3 = Tex("$= 6$",font_size = 80).move_to([3.05,-1,0]).set_opacity(0)
        valor_factorial_4 = Tex("$= 24$",font_size = 80).move_to([4.6,-2,0]).set_opacity(0)
        valor_factorial_5 = Tex("$= 120$",font_size = 80).move_to([6.35,-3,0]).set_opacity(0)
        
        self.numeros.add( valor_factorial_2, valor_factorial_3, valor_factorial_4, valor_factorial_5)
        
        self.play(self.numeros.animate.move_to([.45,-.2,0]).scale(.9))
        self.wait(.5)
        self.play(valor_factorial_2.animate.set_opacity(1))
        self.wait(1)
        self.play(valor_factorial_3.animate.set_opacity(1))
        self.wait(1)
        self.play(valor_factorial_4.animate.set_opacity(1))
        self.wait(1)
        self.play(valor_factorial_5.animate.set_opacity(1))
        self.wait(2)
```

```python
%%manim -qh -v WARNING Video2Escena7Formula

class Video2Escena7Formula(Scene):
    def construct(self):
        self.crear_elementos_iniciales()
        self.crear_formula_permutaciones()
        
        self.mostrar_definicion_permutaciones()
        self.resaltar_permutaciones()
        self.mostrar_permutacion_todos_elementos()
        self.explicar_factorial_cero()
        
        self.ecuacion_final()

    def crear_elementos_iniciales(self):
        cinco = Integer(number=5, font_size=80, color=WHITE).scale(1).move_to([-5.1, 1.3, 0])
        tres = Integer(number=3, font_size=80, color=WHITE).scale(1).move_to([-4.1, 1.3, 0])
        
        borde = Line(start=[-5.1, -0.56, 0], end=[-1.6, -0.56, 0], stroke_width=5)
        borde.z_index = -1
        texto_factorial_5 = Tex("$!$", font_size=80).move_to([-2.8, 0.1, 0]).scale(1.4)
        texto_factorial_2 = Tex("$(\qquad)!$", font_size=72).move_to([-3.3, -1.2, 0]).scale(1.4)
        igual = Tex(r"\text{=}", font_size=80).move_to([-0.9, -0.58, 0]).scale(1.4)
        guion = Tex(r"\text{-}", font_size=80).move_to([-4.6, 1.3, 0])
        
        copia_cinco = cinco.copy().scale(1.4).move_to([-3.3, 0.08, 0])
        copia_cinco_tres = VGroup(cinco.copy(), guion, tres).scale(1.3).move_to([-3.43, -1.2, 0])
        
        self.ecuacion_inicial = VGroup(copia_cinco, copia_cinco_tres, 
                          borde, texto_factorial_5, 
                          texto_factorial_2, igual).move_to([-2.88958072, 0.8, 0])
        
        self.add(self.ecuacion_inicial)
        self.wait(1)

    def crear_formula_permutaciones(self):
        texto_permutaciones = Tex(r"\text{Fórmula de las permutaciones}", 
                                  font_size=80).move_to([-0.1, 2.8, 0])
        self.n_en_n_menos_r = Tex("$\\frac{n!}{(n-r)!}$", 
                                  font_size=80).scale(2).move_to([2.05, 0.75, 0])
        texto_n5 = Tex(r"\text{con n = 5, r = 3}", 
                       font_size=70).move_to([-2.5, -1.3, 0])
        
        self.play(Write(texto_permutaciones))
        self.play(Write(self.n_en_n_menos_r))
        self.play(Write(texto_n5))
        self.wait(3)
        
        self.play(FadeOut(self.ecuacion_inicial, texto_n5))    

    def mostrar_definicion_permutaciones(self):
        self.texto_npk = Tex(r"\[{}_nP_r\]", font_size=80
                            ).move_to([-3.75, 0.9, 0]).scale(1.2)
        texto_igual_p = Tex(r"=", font_size=80).move_to([-1.95, 0.9, 0]).scale(1.2)
        self.texto_definicion = Tex(r"\text{ordenar r elementos de un conjunto} \\ "
                                    r"\text{de n elementos distintos}", font_size=50
                                   ).move_to([0, -1.3, 0])
        
        self.play(
            Write(self.texto_npk), 
            FadeIn(texto_igual_p), 
            self.n_en_n_menos_r.animate.move_to([0.0, 0.88, 0]).scale(0.7)
        )
        self.wait(3)
        
        self.play(Write(self.texto_definicion))
        self.wait(3)

    def resaltar_permutaciones(self):
        self.play(self.texto_npk.animate.scale(1.3).set_color(RED))
        self.wait(3)
        
        self.play(self.texto_npk.animate.scale(0.7692).set_color(WHITE))
        self.wait(1)
        
        self.play(Unwrite(self.texto_definicion))

    def mostrar_permutacion_todos_elementos(self):
        texto_definicion_todos = Tex(r"\text{ordenar todos los n elementos de un conjunto}", 
                                     font_size=50).move_to([0, -1.8, 0])
        formula_nn = Tex("$\\frac{n!}{(n-n)!}$", font_size=80
                        ).move_to([0.0, 0.88, 0]).scale(1.4)
        texto_npn = Tex(r"\[{}_nP_n\]", font_size=80
                       ).move_to([-3.75, 0.9, 0]).scale(1.2)
        
        self.play(Write(texto_definicion_todos))
        self.wait(1.4)
        
        self.play(
            Transform(self.n_en_n_menos_r, formula_nn), 
            Transform(self.texto_npk, texto_npn)
        )
        self.wait(1)

    def explicar_factorial_cero(self):
        corchete = Brace(self.n_en_n_menos_r, DOWN, sharpness=1)
        texto_factorial_0 = Tex("$0!$", font_size=80).next_to(corchete, DOWN)
        igual_1 = Tex("$= 1 $", font_size=80).next_to(texto_factorial_0, RIGHT)
        
        self.play(Write(corchete), Write(texto_factorial_0))
        self.wait(1)
        self.play(Write(igual_1))
        self.wait(3)

    def ecuacion_final(self):
        formula_final = Tex("$= \\frac{n!}{1} = n!$", font_size=80
                           ).scale(1.4).move_to([4.05, 0.88, 0])
        
        self.play(Write(formula_final))
        self.wait(1)
```

```python
%%manim -qh -v WARNING Video2Escena8Ejercicio

class Video2Escena8Ejercicio(Scene):
    def construct(self):
        self.configurar_escena()
        self.mostrar_bus()
        self.transformar_carrusel()
        
        informacion = self.mostrar_textos()
        self.calcular_permutaciones(informacion)

    def configurar_escena(self):
        self.parque = SVGMobject("imagenes/atracciones/parque_diversiones_1.svg")
        self.carrusel = SVGMobject("imagenes/atracciones/carrusel.svg")
        
        self.parque.scale(3.5).move_to([0, 0, -2])
        self.carrusel.scale(1.6).move_to([4.5, -0.83, 0])
        
        self.play(FadeIn(self.parque, self.carrusel), run_time=1)

    def mostrar_bus(self):
        bus = SVGMobject("imagenes/atracciones/bus.svg")
        bus.scale(0.9).move_to([-9.3, -2.5, -2])
        
        self.add(bus)
        self.play(bus.animate.move_to([-3, -2.5, -2]), run_time=2)
        self.wait(3)
        
        self.remove(self.parque, bus)

    def transformar_carrusel(self):
        self.play(self.carrusel.animate.scale(1.8).move_to([0, 0, 0]))
        self.play(self.carrusel.animate.scale(0.6).move_to([-4.5, 1.8, 0]))

    def mostrar_textos(self):
        texto_estudiantes_17 = Tex(r"\text{17 estudiantes}", font_size=50
                                  ).move_to([-0.2, 2.1, 0])
        texto_asientos_5 = Tex(r"\text{5 lugares en el carrusel}", font_size=50
                              ).move_to([0.82, 1.2, 0])
        texto_pregunta = Tex(r"\text{¿De cuántas maneras se pueden subir los} \\ "
                             r"\text{estudiantes en la primer ronda} \\ "
                             r"\text{del carrusel?}", font_size=60)
        
        informacion = VGroup(texto_estudiantes_17, texto_asientos_5)
        
        texto_pregunta.move_to([0, -1.5, 0])
        
        self.play(Write(texto_estudiantes_17))
        self.play(Write(texto_asientos_5))
        self.play(Write(texto_pregunta))
        self.wait(1)
        
        self.play(FadeOut(texto_pregunta))
        self.wait(1)

        return informacion

    def calcular_permutaciones(self, informacion):
        texto_nPkT = Tex(r"\[{}_{17}P_5\]", font_size=80 \
                        ).move_to([-5.25, -0.75, 0]).scale(1.1)
        texto_nPkFT = Tex("$= \\frac{17!}{(17-5)!} = \\frac{17!}{12!}$", \
                          font_size=80).scale(1.1).move_to([-1, -0.75, 0])
        
        texto_n17Fac = Tex("$= \\frac{17 \\times 16 \\times"
                           " \\cdot \\cdot \\cdot \\times 12 \\times 11 \\times "
                           " \\cdot \\cdot \\cdot \\times 2 \\times 1}{12"
                           " \\times 11 \\times \\cdot \\cdot \\cdot \\times 2"
                           " \\times 1}$",font_size = 80)
        texto_n1713Fac = Tex(r"$= 17 \times 16 \times 15 \times 14 \times 13$", 
                             font_size=80)
        resultado = Tex(r"$= 742560$", font_size=80)
        
        linea_1 = Line(start=[-0.96, -2.85, 0], end=[1.6, -1.75, 0], color=RED)
        linea_2 = Line(start=[0.36, -2.85, 0], end=[2.92, -1.75, 0], color=RED)
        linea_3 = Line(start=[2.79, -2.85, 0], end=[5.1, -1.75, 0], color=RED)
        linea_4 = Line(start=[3.72, -2.85, 0], end=[6.08, -1.75, 0], color=RED)
        
        movimientos = VGroup(texto_nPkT, texto_nPkFT, texto_n17Fac, 
                             linea_1, linea_2, linea_3, linea_4)
        
        texto_n17Fac.scale(1.1).move_to([1.06, -2.3, 0])
        texto_n1713Fac.scale(1.1).move_to([1.03, -1, 0])
        resultado.scale(1.1).move_to([-2.11, -2.5, 0])
        
        self.play(Write(texto_nPkT))
        self.play(Write(texto_nPkFT))
        self.play(Write(texto_n17Fac))
        
        self.play(Create(linea_1))
        self.play(Create(linea_2))
        self.play(Create(linea_3))
        self.play(Create(linea_4))
        self.wait(2)

        self.play(
            movimientos.animate.move_to([0.01170331, 1.28429603, 0]), 
            FadeOut(informacion, self.carrusel)
        )

        self.play(Write(texto_n1713Fac), Write(resultado))
        self.remove(texto_n17Fac, texto_n1713Fac, linea_1, linea_2, linea_3, linea_4)
        self.play(resultado.animate.move_to([-2.11, 0.8, 0]))
        self.wait(2)
```
