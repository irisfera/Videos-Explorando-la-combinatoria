```python
%%manim -qh -v WARNING Video2Escena2

class Video2Escena2(Scene):
    def construct(self):
        montaña_rusa, vagon, susana, susana_texto, diana, diana_texto, alex, alex_texto = self.mostrar_escena_inicial()
        self.mostrar_pregunta(montaña_rusa)
        self.animar_niños_a_vagon(vagon, susana, susana_texto, diana, diana_texto, alex, alex_texto)
        self.mostrar_asiento_1(vagon, susana, susana_texto, diana, diana_texto, alex, alex_texto)
        self.iluminar_asientos_2_3()
        self.mostrar_asiento_2(diana, alex)
        self.mostrar_asiento_3(diana)
        arbol_susana = self.crear_arbol_susana(diana, alex)
        self.remove(vagon, susana, diana, alex)
        self.mostrar_titulos_asientos(arbol_susana)
        self.clear()

    def mostrar_escena_inicial(self):
        montaña_rusa = SVGMobject("imagenes/atracciones/montaña_rusa.svg")
        montaña_rusa.scale(2.9).move_to([0, .5, 0])
        montaña_rusa.set_fill(opacity=0.4)
        montaña_rusa.set_stroke(opacity=0.2)

        susana, susana_texto = self.crear_niñx("imagenes/niñxs/niña_verde.svg", [-5, -2, 0], "Susana", [-5, -0.5, 0])
        diana, diana_texto = self.crear_niñx("imagenes/niñxs/niña_naranja.svg", [-3.5, -2, 0], "Diana", [-3.5, -0.5, 0])
        alex, alex_texto = self.crear_niñx("imagenes/niñxs/niño_amarillo.svg", [-2, -2, 0], "Alex", [-2.1, -0.5, 0])
        vagon = SVGMobject("imagenes/vagon_permutaciones/vagon.svg").scale(1).move_to([3, -2, 1])

        self.play(FadeIn(montaña_rusa))
        self.play(FadeIn(*susana, *diana, *alex, vagon), run_time=1)
        self.wait(1)
        
        return montaña_rusa, vagon, susana, susana_texto, diana, diana_texto, alex, alex_texto

    def mostrar_pregunta(self, montaña_rusa):
        pregunta = Tex(
            r"\text{¿De cuántas maneras distintas pueden sentarse} \\ "
            r"\text{en el vagón para probar todos los} \\ "
            r"\text{distintos puntos de vista?}",
            font_size=50
        )
        pregunta.move_to([0, 2.6, 0])
        self.play(Write(pregunta), FadeOut(montaña_rusa))
        self.wait(2)
        self.play(FadeOut(pregunta))

    def animar_niños_a_vagon(self, vagon, susana, susana_texto, diana, diana_texto, alex, alex_texto):
        self.play(
            susana.animate.move_to([-1.6, 3, 0]).scale(0.65),
            susana_texto.animate.move_to([-0.6, 3, 0]).scale(1.2),
            diana.animate.move_to([-1.6, 1.3, 0]).scale(0.65),
            diana_texto.animate.move_to([-0.6, 1.3, 0]).scale(1.2),
            alex.animate.move_to([-1.6, -0.4, 0]).scale(0.65),
            alex_texto.animate.move_to([-0.6, -0.4, 0]).scale(1.2),
            vagon.animate.move_to([0, -2.5, 0]).scale(1.3)
        )
        niñxs_texto = Tex(r"\text{Son 3 amigos y} \\ \text{hay 3 asientos} \\ \text{disponibles.}", font_size=50)
        niñxs_texto.move_to([-4.5, 2.6, 0])
        self.play(Write(niñxs_texto))
        self.wait(0.5)
        self.play(FadeOut(niñxs_texto))
        
    def mostrar_asiento_1(self, vagon, susana, susana_texto, diana, diana_texto, alex, alex_texto):
        vagon_1 = SVGMobject("imagenes/vagon_permutaciones/vagonL1.svg").move_to([0,-2.5,0]).scale(1.358).set_fill(opacity=1)
        numero_1 = Integer(number=1).scale(2).move_to([-1.5,-1.8,0]).set_color(BLACK) 
        asiento_1_texto = Tex(r"\text{3 personas} \\ \text{elegibles para} \\ \text{el asiento 1.}", font_size=50)
        asiento_1_texto.move_to([-4.5, 2.6, 0])
        self.add(vagon_1,numero_1)
        self.play(Write(asiento_1_texto))
        
        self.wait(.5)
        self.remove(vagon_1, numero_1)
        self.add(vagon)
        self.play(FadeOut(diana_texto, alex_texto))
        self.wait(.5)
        self.play(FadeOut(asiento_1_texto), FadeOut(susana_texto),
                  diana.animate.move_to([.3,2.6,0]).scale(1.4),
                  alex.animate.move_to([.3,0,0]).scale(1.3), 
                  susana.animate.move_to([-1.6,-1.5,-1]).scale(1.4))
        self.wait(.5)
        
    def iluminar_asientos_2_3(self):
        vagon_2_3 = SVGMobject("imagenes/vagon_permutaciones/vagonL2L3.svg").move_to([0,-2.5,0]).scale(1.363).set_fill(opacity=1)
        self.add(vagon_2_3)
        asiento_2_3_texto = Tex(r"\text{Quedan 2 personas y } \\ \text{2 asientos disponibles.}", font_size=50)
        asiento_2_3_texto.move_to([-4, 2.6, 0])
        self.play(Write(asiento_2_3_texto))
        self.wait(.5)
        self.remove(vagon_2_3)
        self.play(FadeOut(asiento_2_3_texto))
        

    def mostrar_asiento_2(self, diana, alex):
        vagon_2 = SVGMobject("imagenes/vagon_permutaciones/vagonL2.svg").move_to([0,-2.5,0]).scale(1.363).set_fill(opacity=1)
        numero_2 = Integer(number=2).scale(2).move_to([.4,-1.8,0]).set_color(BLACK)
        asiento_2_texto = Tex(r"\text{2 personas} \\ \text{elegibles para} \\ \text{el asiento 2.}", font_size=50)
        asiento_2_texto.move_to([-4.5, 2.6, 0])
        self.add(vagon_2, numero_2)
        self.play(Write(asiento_2_texto))
        self.wait(1)
        self.remove(vagon_2, numero_2)
        self.play(FadeOut(asiento_2_texto), diana.animate.move_to([2.1,.5,0]).scale(1),
                  alex.animate.move_to([.3,-1.7,0]).scale(1.1))
        self.wait(.5)
    
    def mostrar_asiento_3(self, diana):
        vagon_3 = SVGMobject("imagenes/vagon_permutaciones/vagonL3.svg").move_to([0,-2.5,0]).scale(1.347).set_fill(opacity=1)
        numero_3 = Integer(number=3).scale(2).move_to([2.2,-1.8,0]).set_color(BLACK)
        asiento_3_texto = Tex(r"\text{1 personas} \\ \text{elegibles para} \\ \text{el asiento 3.}", font_size=50)
        asiento_3_texto.move_to([-4.5, 2.6, 0])
        self.add(vagon_3, numero_3)
        self.play(Write(asiento_3_texto))
        self.wait(1)
        self.remove(vagon_3, numero_3)
        self.play(FadeOut(asiento_3_texto), diana.animate.move_to([2.1,-1.7,0]).scale(1.1))
        self.wait(1.2)
        
    def crear_arbol_susana(self, diana, alex):
        arbol_susana = VGroup()
        susana_1 = SVGMobject("imagenes/niñxs/niña_verde.svg").scale(0.8).move_to([-4, 2, 0])
        self.arbol_parte_1(susana_1, arbol_susana)
        self.cambiar_diana_alex(diana, alex)
        self.arbol_parte_2(susana_1, arbol_susana)
        
        return arbol_susana

    def arbol_parte_1(self, susana_1, arbol_susana):
        alex_1 = SVGMobject("imagenes/niñxs/niño_amarillo.svg").scale(0.8).move_to([-1, 3, 0])
        diana_1 = SVGMobject("imagenes/niñxs/niña_naranja.svg").scale(0.85).move_to([2, 3, 0])
        vagon_1_3_2 = SVGMobject("imagenes/vagon_permutaciones/vagon132.svg").scale(0.7).move_to([5, 3, 0])

        arista_1 = self.crear_arista(susana_1, alex_1)
        arista_2 = self.crear_arista(alex_1, diana_1)
        arista_3 = self.crear_arista(diana_1, vagon_1_3_2)
        
        self.play(FadeIn(susana_1))
        self.play(Create(arista_1), FadeIn(alex_1))
        self.play(Create(arista_2), FadeIn(diana_1))
        self.play(Create(arista_3), FadeIn(vagon_1_3_2))
        
        arbol_susana.add(
            susana_1, diana_1, alex_1,
            arista_1, arista_2, arista_3,
            vagon_1_3_2
        )
        

    def cambiar_diana_alex(self, diana, alex):
        self.play(
            diana.animate.move_to([2.1, 0.5, 0]),
            alex.animate.move_to([0.3, 0, 0])
        )
        self.play(
            diana.animate.move_to([0.34, -1.5, 0]),
            alex.animate.move_to([2.2, -1.79, 0])
        )
    
    def arbol_parte_2(self, susana_1, arbol_susana):
        alex_2 = SVGMobject("imagenes/niñxs/niño_amarillo.svg").scale(0.8).move_to([2, 1, 0])
        diana_2 = SVGMobject("imagenes/niñxs/niña_naranja.svg").scale(0.85).move_to([-1, 1, 0])
        vagon_1_2_3 = SVGMobject("imagenes/vagon_permutaciones/vagon123.svg").scale(0.7).move_to([5, 1, 0])

        arista_4 = self.crear_arista(susana_1, diana_2)
        arista_5 = self.crear_arista(diana_2, alex_2)
        arista_6 = self.crear_arista(alex_2, vagon_1_2_3)
        
        arbol_susana.add(alex_2, diana_2, arista_4, arista_5, arista_6, vagon_1_2_3)
        
        self.play(Create(arista_4), FadeIn(diana_2))
        self.play(Create(arista_5), FadeIn(alex_2))
        self.play(Create(arista_6), FadeIn(vagon_1_2_3))

    def mostrar_titulos_asientos(self, arbol_susana):
        asiento_1_texto = Tex(r"\text{Asiento} \\ \text{1}", font_size=50).move_to([-5.35, 2.6, 0])
        asiento_2_texto = Tex(r"\text{Asiento} \\ \text{2}", font_size=50).move_to([-2.25, 2.6, 0])
        asiento_3_texto = Tex(r"\text{Asiento} \\ \text{3}", font_size=50).move_to([1.05, 2.6, 0])
        self.play(
            arbol_susana.animate.move_to([0, -1, 0]).scale(1.07),
            FadeIn(asiento_1_texto, asiento_2_texto, asiento_3_texto)
        )
        self.wait(2)
        

    def crear_niñx(self, svg_path, pos, nombre, text_pos):
        niñx = SVGMobject(svg_path).scale(1.2).move_to(pos)
        texto = Tex(nombre, font_size=20).move_to(text_pos).scale(1.2)
        return niñx, texto

    def crear_arista(self, start_obj, end_obj):
        linea = Line(start=start_obj.get_center(), end=end_obj.get_center()).set_z_index(-1)
        #edge.z_index = -1
        return linea
```

```python
%%manim -qh -v WARNING Video2Escena3

class Video2Escena3(Scene):
    def construct(self):
        asiento_1_texto, asiento_2_texto, asiento_3_texto = self.mostrar_asientos()
        arbol = self.crear_arbol_con_asientos()
        arbol.add(asiento_1_texto, asiento_2_texto, asiento_3_texto)
        self.play(arbol.animate.scale(.95).move_to([1.2, -.3, 0]))
        self.mostrar_principio_producto()
        self.mostrar_operaciones(arbol)
        self.animar_movimientos_asientos(asiento_1_texto, asiento_2_texto, asiento_3_texto)
        self.esperar_y_despedir(arbol)

    def mostrar_asientos(self):
        asiento_1_texto = self.crear_asiento_texto(1, [-2.1, 3, 0])
        asiento_2_texto = self.crear_asiento_texto(2, [0.1, 3, 0])
        asiento_3_texto = self.crear_asiento_texto(3, [2.3, 3, 0])
        
        self.play(FadeIn(asiento_1_texto, asiento_2_texto, asiento_3_texto))
        
        return asiento_1_texto, asiento_2_texto, asiento_3_texto

    def crear_asiento_texto(self, numero, posicion):
        return Tex(rf"\text{{Asiento}} \\ \text{{{numero}}}", font_size=50).move_to(posicion)


    def crear_arbol_con_asientos(self):
        arbol_niveles, arbol, coordenadas,aristas = construir_arbol_vagon(1)
        arbol.move_to([0, -0.8, 0])
        crear_arbol_niveles(self, arbol_niveles)
        self.wait(2)        
        return arbol


    def mostrar_principio_producto(self):
        regla_producto_texto = Tex(r"\text{Principio del} \\ \text{Producto}", font_size=60)
        regla_producto_texto.move_to([-4.5, 3, 0])
        self.play(Write(regla_producto_texto))

    def mostrar_operaciones(self, arbol):
        self.mostrar_ovalos_y_numeros(arbol)

    def mostrar_ovalos_y_numeros(self, arbol):
        ovalo1 = self.crear_ovalo([-.78, -0.9, 0])
        self.play(Write(ovalo1))
        numero1 = Integer(number=3, color=WHITE).scale(1.5).move_to([-6, 1, 0])
        self.play(Transform(ovalo1, numero1))
        self.mostrar_multiplicacion([-5.5, 1, 0])

        ovalo2 = self.crear_ovalo([1.3, -0.95, 0], 1.9)
        self.play(Write(ovalo2))
        numero2 = Integer(number=2, color=WHITE).scale(1.5).move_to([-5, 1, 0])
        self.play(Transform(ovalo2, numero2))
        self.mostrar_multiplicacion([-4.5, 1, 0])

        ovalo3 = self.crear_ovalo([3.4, -0.46, 0], 1)
        self.play(Write(ovalo3))
        numero3 = Integer(number=1, color=WHITE).scale(1.5).move_to([-4, 1, 0])
        self.play(Transform(ovalo3, numero3))
        
        resultado = Tex("= 6", font_size=80).next_to([-4, 1, 0], RIGHT)
        self.play(FadeIn(resultado))

    def crear_ovalo(self, posicion, altura=5.2):
        return Ellipse(width=1.5, height=altura, color=BLUE, fill_opacity=0, stroke_width=5).move_to(posicion)

    def mostrar_multiplicacion(self, posicion):
        multiplicador = Tex("x", font_size=80).move_to(posicion)
        self.play(FadeIn(multiplicador))

    def animar_movimientos_asientos(self, asiento_1_texto, asiento_2_texto, asiento_3_texto):
        posiciones = [
            [1.295, 2.662, 0],
            [-0.795, 2.662, 0],
            [3.385, 2.662, 0]
        ]

        self.play(
            asiento_1_texto.animate.move_to(posiciones[0]),
            asiento_2_texto.animate.move_to(posiciones[1])
        )
        self.play(
            asiento_1_texto.animate.move_to(posiciones[2]),
            asiento_3_texto.animate.move_to(posiciones[0])
        )
        self.play(
            asiento_3_texto.animate.move_to(posiciones[1]),
            asiento_1_texto.animate.move_to(posiciones[0]),
            asiento_2_texto.animate.move_to(posiciones[2])
        )

    def obtener_asientos_texto(self):
        asiento_1_texto = self.crear_asiento_texto("Asiento 1", [-2.1, 3, 0])
        asiento_2_texto = self.crear_asiento_texto("Asiento 2", [0.1, 3, 0])
        asiento_3_texto = self.crear_asiento_texto("Asiento 3", [2.3, 3, 0])
        return asiento_1_texto, asiento_2_texto, asiento_3_texto

    def esperar_y_despedir(self, arbol):
        self.wait(2)
        self.play(FadeOut(arbol))
        self.wait(2)
```

```python
%%manim -qh -v WARNING Video2Escena4

class Video2Escena4(Scene):
    def construct(self):
        regla_prod_texto, factorial_3_grupo, igual_6 = self.mostrar_escena_pasada()
        self.factorial_3(regla_prod_texto, factorial_3_grupo, igual_6)
        factoriales, n_factorial = self.factorial()
        igual_2, factorial_1_igual = self.cambiar_opacidad_2(factoriales)
        igual_3, factorial_2_igual = self.cambiar_opacidad_3(factoriales)
        igual_4, factorial_3_igual = self.cambiar_opacidad_4(factoriales)
        igual_5, factorial_4_igual = self.cambiar_opacidad_5(factoriales)
        numeros = self.opacidad_1(igual_2, factorial_1_igual, igual_3, factorial_2_igual, igual_4, 
                        factorial_3_igual,factorial_4_igual, factoriales)
        numeros.add(igual_5, n_factorial)
        for factorial in factoriales:
            numeros.add(factorial)
        
        self.igualdad(numeros)
        self.clear()
        
    def mostrar_escena_pasada(self):
        regla_prod_texto = Tex(r"\text{Principio del} \\ \text{Producto}",font_size = 60).move_to([-4.5, 3, 0])
        numero_3 = Integer(number=3, color = WHITE).scale(1.5).move_to([-6,1,0])
        multiplicador_1 = Tex("$\\times$",font_size = 80).move_to([-5.5,1,0])
        numero_2 = Integer(number=2, color = WHITE).scale(1.5).move_to([-5,1,0])
        multiplicador_2 = Tex("$\\times$",font_size = 80).move_to([-4.5,1,0])
        numero_1 = Integer(number=1, color = WHITE).scale(1.5).move_to([-4,1,0])
        igual_6 = Tex("$= 6$",font_size = 80).next_to(numero_1, RIGHT)
        
        factorial_3_grupo = VGroup()
        factorial_3_grupo.add(numero_3, numero_2, numero_1, multiplicador_1, multiplicador_2)
        
        self.add(regla_prod_texto, numero_3, numero_2, numero_1, multiplicador_1, multiplicador_2, igual_6)
        self.wait(2)
        
        return regla_prod_texto, factorial_3_grupo, igual_6
    
    def factorial_3(self, regla_prod_texto, factorial_3_grupo, igual_6):
        igual = Tex("$=$",font_size = 80).move_to([-1.8,.94,0])
        factorial_3 = Tex("$3!$",font_size = 80).move_to([-1.0,1,0])
        self.play(FadeIn(igual, factorial_3))
        self.wait(2)
        factorial_texto = Tex(r"\text{Factorial}",font_size = 60).move_to([0,3,0])
        flecha = Arrow([-.77, 2.95, 0], [-.77, 1.1, 0])
        self.play(Write(factorial_texto), Write(flecha))
        self.wait(2)
        self.play(FadeOut(regla_prod_texto, igual_6))
        self.play(igual.animate.move_to([-.16,1,0]), factorial_3_grupo.animate.move_to([1.45,1,0]))
        factorial_3_grupo.add(factorial_3, igual)
        self.wait(2)
        self.play(factorial_texto.animate.scale(1.3).move_to([-4.5,3,0]), FadeOut(flecha, factorial_3_grupo))
        self.wait(2)
        
    def factorial(self):
        n_factorial = Tex("$n! = n \\times (n-1) \\times \\cdot \\cdot \\cdot \\times 2 \\times 1$",font_size = 80).move_to([-0.8,2,0])
        self.play(Write(n_factorial))
        factoriales = self.crear_factoriales()
        for factorial in factoriales:
            self.play(Write(factorial))
            self.wait(.5)
        self.wait(1)
        
        return factoriales, n_factorial
        
    def crear_factoriales(self):
        factoriales = [
            Tex("$1! = 1$", font_size=80).move_to([-4.325, 1, 0]),
            Tex("$2! = 2 \\times 1$", font_size=80).move_to([-3.625, 0, 0]),
            Tex("$3! = 3 \\times 2 \\times 1$", font_size=80).move_to([-2.911, -1, 0]),
            Tex("$4! = 4 \\times 3 \\times 2 \\times 1$", font_size=80).move_to([-2.205, -2, 0]),
            Tex("$5! = 5 \\times 4 \\times 3 \\times 2 \\times 1$", font_size=80).move_to([-1.48, -3, 0])
        ]
        return factoriales
    
    def cambiar_opacidad_2(self, factoriales):
        igual_2 = Tex("$= 2 \\times$",font_size = 80).move_to([-.8,0,0])
        factorial_1 = Tex("$1!$",font_size = 80).move_to([.5,0,0])
        self.cambiar_opactidad(factoriales, 1)
        factorial_1_igual = copy.deepcopy(factoriales[0])
        self.add(factorial_1_igual)
        self.play(FadeIn(igual_2))
        self.play(Transform(factorial_1_igual, factorial_1), run_time = 1)
        self.wait(1)
        self.play(factoriales[1].animate.set_opacity(.2),
                  factorial_1_igual.animate.set_opacity(.1),
                  igual_2.animate.set_opacity(.2),
                  factoriales[2].animate.set_opacity(1))
        
        return igual_2, factorial_1_igual
        
    def cambiar_opacidad_3(self, factoriales):
        factorial_2_igual = copy.deepcopy(factoriales[1])
        igual_3 = Tex("$= 3 \\times$",font_size = 80).move_to([.6,-1,0])
        factorial_2 = Tex("$2!$",font_size = 80).move_to([1.9,-1,0])
        self.play(FadeIn(igual_3))
        self.play(Transform(factorial_2_igual, factorial_2), run_time = 1)
        self.wait(1)
        self.play(factoriales[2].animate.set_opacity(.2),
                  factorial_2_igual.animate.set_opacity(.075),
                  igual_3.animate.set_opacity(.2),
                  factoriales[3].animate.set_opacity(1))
        return igual_3, factorial_2_igual
        
    def cambiar_opacidad_4(self, factoriales):
        factorial_3_igual = copy.deepcopy(factoriales[2])
        igual_4 = Tex("$= 4 \\times$",font_size = 80).move_to([2,-2,0])
        factorial_3 = Tex("$3!$",font_size = 80).move_to([3.3,-2,0])
        self.play(FadeIn(igual_4))
        self.play(Transform(factorial_3_igual, factorial_3), run_time = 1)
        self.wait(1)
        self.play(factoriales[3].animate.set_opacity(.2),
                  factorial_3_igual.animate.set_opacity(.05),
                  igual_4.animate.set_opacity(.2),
                  factoriales[4].animate.set_opacity(1))
        return igual_4, factorial_3_igual
        
    def cambiar_opacidad_5(self, factoriales):
        factorial_4_igual = copy.deepcopy(factoriales[3])
        igual_5 = Tex("$= 5 \\times$",font_size = 80).move_to([3.5,-3,0])
        factorial_4 = Tex("$4!$",font_size = 80).move_to([4.8,-3,0])
        self.play(FadeIn(igual_5))
        self.play(Transform(factorial_4_igual, factorial_4), run_time = 1)
        self.wait(2)
        return igual_5, factorial_4_igual
        
    def cambiar_opactidad(self, factoriales, indice):
        self.play(*[factorial.animate.set_opacity(.2) for i, factorial in enumerate(factoriales) if i != indice])
        
    def opacidad_1(self, igual_2, factorial_1_igual, igual_3, factorial_2_igual, igual_4, factorial_3_igual,
                   factorial_4_igual, factoriales):
        
        self.play(*[factorial.animate.set_opacity(1) for i, factorial in enumerate(factoriales) if i != 5],

                  factorial_1_igual.animate.set_opacity(1),
                  igual_2.animate.set_opacity(1),
                  factorial_2_igual.animate.set_opacity(1),
                  igual_3.animate.set_opacity(1),
                  factorial_3_igual.animate.set_opacity(1),
                  igual_4.animate.set_opacity(1),
                  factorial_4_igual.animate.set_opacity(1),
            )
        numeros = VGroup(igual_2, factorial_1_igual, igual_3, factorial_2_igual, igual_4, factorial_3_igual,
                   factorial_4_igual)
        
        return numeros

    def igualdad(self, numeros):
        valor_factorial_2 = Tex("$= 2$",font_size = 80).move_to([1.65,0,0]).set_opacity(0)
        valor_factorial_3 = Tex("$= 6$",font_size = 80).move_to([3.05,-1,0]).set_opacity(0)
        valor_factorial_4 = Tex("$= 24$",font_size = 80).move_to([4.6,-2,0]).set_opacity(0)
        valor_factorial_5 = Tex("$= 120$",font_size = 80).move_to([6.35,-3,0]).set_opacity(0)
        
        numeros.add( valor_factorial_2, valor_factorial_3, valor_factorial_4, valor_factorial_5)
        self.play(numeros.animate.move_to([.45,-.2,0]).scale(.9))
        self.wait(.5)
        self.play(valor_factorial_2.animate.set_opacity(1))
        self.wait(1)
        self.play(valor_factorial_3.animate.set_opacity(1))
        self.wait(1)
        self.play(valor_factorial_4.animate.set_opacity(1))
        self.wait(1)
        self.play(valor_factorial_5.animate.set_opacity(1))
        self.wait(2)
```

```python
%%manim -qh -v WARNING Video2Escena7

class Video2Escena7(Scene):
    def construct(self):
        ecuacion_inicial = self.crear_elementos_iniciales()
        n_en_n_menos_r = self.crear_formula_permutaciones(ecuacion_inicial)
        texto_npk, texto_definicion = self.mostrar_definicion_permutaciones(n_en_n_menos_r)
        self.resaltar_permutaciones(texto_npk, texto_definicion)
        self.mostrar_permutacion_todos_elementos(n_en_n_menos_r, texto_npk)
        self.explicar_factorial_cero( n_en_n_menos_r)
        self.ecuacion_final()

    def crear_elementos_iniciales(self):
        cinco = Integer(number=5, font_size=80, color=WHITE).scale(1).move_to([-5.1, 1.3, 0])
        tres = Integer(number=3, font_size=80, color=WHITE).scale(1).move_to([-4.1, 1.3, 0])
        guion = Tex(r"\text{-}", font_size=80).move_to([-4.6, 1.3, 0])
        
        borde = Line(start=[-5.1, -0.56, 0], end=[-1.6, -0.56, 0], stroke_width=5)
        borde.z_index = -1
        texto_factorial_5 = Tex("$!$", font_size=80).move_to([-2.8, 0.1, 0]).scale(1.4)
        texto_factorial_2 = Tex("$(\qquad)!$", font_size=72).move_to([-3.3, -1.2, 0]).scale(1.4)
        igual = Tex(r"\text{=}", font_size=80).move_to([-0.9, -0.58, 0]).scale(1.4)
        
        copia_cinco = cinco.copy().scale(1.4).move_to([-3.3, 0.08, 0])
        copia_cinco_tres = VGroup(cinco.copy(), guion, tres).scale(1.3).move_to([-3.43, -1.2, 0])
        
        ecuacion = VGroup(copia_cinco, copia_cinco_tres, borde, texto_factorial_5, texto_factorial_2, igual).move_to([-2.88958072, 0.8, 0])
        
        self.add(ecuacion)
        self.wait(1)
        
        return ecuacion

    def crear_formula_permutaciones(self, ecuacion_inicial):
        texto_permutaciones = Tex(r"\text{Fórmula de las permutaciones}", font_size=80).move_to([-0.1, 2.8, 0])
        self.play(Write(texto_permutaciones))
        
        n_en_n_menos_r = Tex("$\\frac{n!}{(n-r)!}$", font_size=80).scale(2).move_to([2.05, 0.75, 0])
        self.play(Write(n_en_n_menos_r))
        
        texto_n5 = Tex(r"\text{con n = 5, r = 3}", font_size=70).move_to([-2.5, -1.3, 0])
        self.play(Write(texto_n5))
        
        self.wait(3)
        
        self.play(FadeOut(ecuacion_inicial, texto_n5))
        
        return n_en_n_menos_r
        

    def mostrar_definicion_permutaciones(self, n_en_n_menos_r):
        texto_npk = Tex(r"\[{}_nP_r\]", font_size=80).move_to([-3.75, 0.9, 0]).scale(1.2)
        texto_igual_p = Tex(r"=", font_size=80).move_to([-1.95, 0.9, 0]).scale(1.2)
        
        self.play(Write(texto_npk), FadeIn(texto_igual_p), n_en_n_menos_r.animate.move_to([0.0, 0.88, 0]).scale(0.7))
        self.wait(3)
        
        texto_definicion = Tex(r"\text{ordenar r elementos de un conjunto} \\ \text{de n elementos distintos}", font_size=50).move_to([0, -1.3, 0])
        self.play(Write(texto_definicion))
        self.wait(3)
        
        return texto_npk, texto_definicion

    def resaltar_permutaciones(self, texto_npk, texto_definicion):
        self.play(texto_npk.animate.scale(1.3).set_color(RED))
        self.wait(3)
        self.play(texto_npk.animate.scale(0.7692).set_color(WHITE))
        self.wait(1)
        self.play(Unwrite(texto_definicion))

    def mostrar_permutacion_todos_elementos(self, n_en_n_menos_r, texto_npk):
        texto_definicion_todos = Tex(r"\text{ordenar todos los n elementos de un conjunto}", font_size=50).move_to([0, -1.8, 0])
        self.play(Write(texto_definicion_todos))
        self.wait(1.4)
        
        formula_nn = Tex("$\\frac{n!}{(n-n)!}$", font_size=80).move_to([0.0, 0.88, 0]).scale(1.4)
        texto_npn = Tex(r"\[{}_nP_n\]", font_size=80).move_to([-3.75, 0.9, 0]).scale(1.2)
        
        self.play(Transform(n_en_n_menos_r, formula_nn), Transform(texto_npk, texto_npn))
        self.wait(1)

    def explicar_factorial_cero(self,  n_en_n_menos_r):
        corchete = Brace( n_en_n_menos_r, DOWN, sharpness=1)
        texto_factorial_0 = Tex("$0!$", font_size=80).next_to(corchete, DOWN)
        self.play(Write(corchete), Write(texto_factorial_0))
        self.wait(1)
        
        igual_1 = Tex("$= 1 $", font_size=80).next_to(texto_factorial_0, RIGHT)
        self.play(Write(igual_1))
        self.wait(3)

    def ecuacion_final(self):
        formula_final = Tex("$= \\frac{n!}{1} = n!$", font_size=80).scale(1.4).move_to([4.05, 0.88, 0])
        self.play(Write(formula_final))
        self.wait(1)
```
