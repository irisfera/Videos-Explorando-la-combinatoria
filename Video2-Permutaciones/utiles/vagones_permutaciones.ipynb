```python
def distribuir_puntos_uniformemente(n, distancia):
    puntos = []
    if n % 2 == 0:  # Si n es par, se divide niformemente a ambos lados de 0.
        cant_puntos_lado = n // 2
        puntos_lado = [distancia * (i - 1/2) for i in range(1, cant_puntos_lado + 1)]
    else:  # Si n es impar, se coloca un punto en 0 y se divide el resto uniformemente.
        cant_puntos_lado = (n - 1) // 2
        puntos_lado = [distancia * i for i in range(1, cant_puntos_lado + 1)]
        puntos += [0]
    
    # Distribuye puntos uniformemente arriba y abajo de 0 con una distancia de dist.
    puntos.extend(puntos_lado)
    puntos.extend(-1* np.array(puntos_lado))

    return sorted(puntos, reverse=True)
```

```python
def construir_arbol_vagon(distancia = .8):
    dec_1 = 3
    dec_2 = 2
    dec_3 = 1
    valores = [dec_1, dec_2, dec_3, 1]
    coordenadas, aristas = multiplicar_valores(valores, distancia)
    
    combinaciones = matriz_permutaciones_niñxs()
    
    arbol_niveles, arbol, aristas = construir_pisos_arbol_vagon(coordenadas, combinaciones, aristas, valores)
            
    return (arbol_niveles, arbol, coordenadas, aristas)

def construir_pisos_arbol_vagon(coordenadas, combinaciones, aristas, valores):
    arbol_niveles = []
    arbol = VGroup()
    i = 0
    j=0
    k = 0
    num_aristas = 1
    for piso in coordenadas:
        nivel = VGroup()
        for punto in piso:
            escala = .4
            #if k == 1:
            #    escala = .4
            combinaciones[i].move_to(punto).scale(escala)
            arbol.add(combinaciones[i])
            nivel.add(combinaciones[i])
            i +=1 
        if k < len(valores):
            num_aristas *= valores[k]
            r = 0
            while j < len(aristas) and r < num_aristas:
                linea = Line(start = aristas[j][0], end = aristas[j][1])
                linea.z_index = -5
                #self.play(Write(edge))
                arbol.add(linea)
                nivel.add(linea)
                j += 1
                r += 1
            
        arbol_niveles.append(nivel)
        k += 1
        
    return arbol_niveles, arbol, aristas

    
```

```python
def crear_arbol_niveles(self,arbol_niveles):
    for obj in arbol_niveles:
        group = VGroup(*obj)
        self.add(group)
        self.play(Write(group))
```
