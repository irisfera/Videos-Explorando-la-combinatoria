```python
from manim import *
import networkx as nx
import numpy as np
import math
```

```python
# Restaurante
def restaurante(self):
        r = SVGMobject("imagenes/comida/restaurante.svg")
        r.scale(3.3)
        restaurant = Text('IRISFERA', color=BLACK)
        restaurant.move_to([-0.1, 1.5, 0]).scale(1.1)
        self.play(Write(r), run_time=2.5)
        self.play(Write(restaurant))
        self.play(restaurant.animate.move_to([-13, 17, 0]).scale(6), 
                  FadeOut(r, target_position= [-12, 5, 0], scale=7.5), 
                  run_time=1.3)
```

```python
def tipos_pasta(self):
    # Crea el primer bloque de texto "Tipos de pasta"
    texto_tipos = Tex("3 tipos").next_to(ORIGIN, LEFT)
    texto_pasta = Tex("de pasta").next_to(texto_tipos, RIGHT)

    #Agregar tipos de pasta
    espagueti = SVGMobject("imagenes/comida/espagueti.svg")
    espagueti.move_to((np.array([-5, 2, 0]))).scale(1.5)

    fusili = SVGMobject("imagenes/comida/fusili.svg")
    fusili.move_to((np.array([-2, 2, 0]))).scale(1)

    macaroni = SVGMobject("imagenes/comida/macaroni.svg")
    macaroni.move_to((np.array([1, 2, 0]))).scale(1)

    self.play(Write(texto_tipos))
    self.play(Write(texto_pasta))

    self.play(
        texto_tipos.animate.move_to([5, 2.5, 0]),
        texto_pasta.animate.next_to([5, 2.5, 0], DOWN),
        Write(espagueti), run_time = .4
    )

    self.play(Write(fusili), run_time=.4)
    self.play(Write(macaroni), run_time=.4)
    
    return (texto_tipos, texto_pasta, espagueti, fusili, macaroni)
```

```python
def tipos_pasta_2(self):
    # Crea el primer bloque de texto "Tipos de pasta"
    texto_tipos = Tex("2 tipos").next_to(ORIGIN, LEFT)
    texto_pasta = Tex("de pasta").next_to(texto_tipos, RIGHT)

    #Agregar tipos de pasta
    espagueti = SVGMobject("imagenes/comida/espagueti.svg")
    espagueti.move_to((np.array([-5, 2, 0]))).scale(.95).rotate(-.3)

    fusili = SVGMobject("imagenes/comida/fusili.svg")
    fusili.move_to((np.array([-3, 2, 0]))).scale(.68)

    self.play(Write(texto_tipos))
    self.play(Write(texto_pasta))

    
    self.play(
        texto_tipos.animate.move_to([-5, 3.5, 0]),
        texto_pasta.animate.move_to([-3.2, 3.5, 0]),
        Write(espagueti), run_time = .4
    )
    self.play(Write(fusili), run_time=.4)
    return (texto_tipos, texto_pasta, espagueti, fusili)
```

```python
def tipos_salsa(self):
    # Crea el segundo bloque de texto "Tipos de salsa"
    texto_tipos = Tex("3 tipos").next_to(ORIGIN, LEFT)
    texto_salsa = Tex("de salsa").next_to(texto_tipos, RIGHT)

    # Tipos de salsa
    salsa_pomodoro = SVGMobject("imagenes/comida/salsa_pomodoro.svg")
    salsa_pomodoro.move_to((np.array([-5, -2, 0]))).scale(1.5)

    salsa_parmesano = SVGMobject("imagenes/comida/salsa_parmesano.svg")
    salsa_parmesano.move_to((np.array([-2, -2, 0]))).scale(1.5)

    salsa_ajo = SVGMobject("imagenes/comida/salsa_ajo.svg")
    salsa_ajo.move_to((np.array([1, -2, 0]))).scale(1.5)

    self.play(Write(texto_tipos))
    self.play(Write(texto_salsa))

    self.play(
        texto_salsa.animate.move_to([5, -1.5, 0]),
        texto_tipos.animate.next_to([5, -1.5, 0], UP),
        Write(salsa_pomodoro), run_time = .6
    )

    self.play(Write(salsa_parmesano), run_time=.6)
    self.play(Write(salsa_ajo), run_time=1)
    
    return (texto_tipos, texto_salsa, salsa_pomodoro, salsa_parmesano, salsa_ajo)
```

```python
def tipos_salsa_2(self):
    # Crea el segundo bloque de texto "Tipos de salsa"
    texto_tipos = Tex("4 tipos").next_to(ORIGIN, LEFT)
    texto_salsa = Tex("de salsa").next_to(texto_tipos, RIGHT)

    # Tipos de salsa
    salsa_pomodoro = SVGMobject("imagenes/comida/salsa_pomodoro.svg")
    salsa_pomodoro.move_to((np.array([-5.5, -.5, 0]))).scale(.9)

    salsa_parmesano = SVGMobject("imagenes/comida/salsa_parmesano.svg")
    salsa_parmesano.move_to((np.array([-4, -.5, 0]))).scale(.9)
    
    salsa_pesto = SVGMobject("imagenes/comida/salsa_pesto.svg")
    salsa_pesto.move_to((np.array([-2.5, -.5, 0]))).scale(.9)

    salsa_champiñon = SVGMobject("imagenes/comida/salsa_champiñon.svg")
    salsa_champiñon.move_to((np.array([-1, -.5, 0]))).scale(.9)

    self.play(Write(texto_tipos))
    self.play(Write(texto_salsa))

    self.play(
        texto_salsa.animate.move_to([-3.2, .8, 0]),
        texto_tipos.animate.next_to([-6, .8, 0]),
        Write(salsa_pomodoro), run_time = .8
    )

    self.play(Write(salsa_parmesano), run_time=.8)
    self.play(Write(salsa_pesto), run_time=.8)
    self.play(Write(salsa_champiñon), run_time=1)
    return (texto_tipos, texto_salsa, salsa_pomodoro, salsa_parmesano, salsa_pesto, salsa_champiñon)
```

```python
def tipos_proteina(self):
    # Crea el primer bloque de texto "Tipos de pasta"
    texto_tipos = Tex("2 tipos").move_to([2, 3.5, 0])
    texto_proteina = Tex("de proteina").next_to(texto_tipos, RIGHT)
    
    #Agregar tipos de proteina
    salmon = SVGMobject("imagenes/comida/salmon.svg")
    salmon.move_to((np.array([-5, -3, 0]))).scale(.5)

    carne = SVGMobject("imagenes/comida/carne.svg")
    carne.move_to((np.array([-2, -3, 0]))).scale(.5)

    self.play(Write(texto_tipos))
    self.play(Write(texto_proteina))

    self.play(
        texto_tipos.animate.move_to([-5, -1.8, 0]),
        texto_proteina.animate.next_to([-4.2, -1.8, 0], RIGHT),
        Write(salmon), run_time = .7
    )
    
    self.play(Write(carne), run_time=.6)
    
    return (texto_tipos, texto_proteina, salmon, carne)
```
